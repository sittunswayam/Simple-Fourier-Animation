```python
# Library Imports

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from celluloid import Camera
```

```python
# Defining x axis and y = f(x)

x = np.linspace(0,1,500)

y = signal.square(2 * np.pi * 5 * x)
#y = signal.sawtooth(2 * np.pi * 5 * x)
#y = np.sin(2*np.pi*x) + np.random.normal(scale=0.3, size=len(x))
#y = np.cos(4*np.pi*x) + np.random.normal(scale=0.15, size=len(x))

n = len(y)
plt.plot(x,y)
```

```python
# Plotting the generated figure with n = 0 to 300 coefficients and generating the animated GIF

fig = plt.figure(figsize=(10,10))
camera = Camera(fig)
for i in range(300):
    Y = np.fft.fft(y)/n*1000
    np.put(Y, range(i+1, n), 0.0)
    ifft = np.fft.ifft(Y)
    t = plt.plot(x,ifft,color='r',linewidth=1.5)
    plt.plot(x,y,color='k')
    plt.legend(t, [f'n = {i}'], loc=3)
    plt.grid(True,axis='both')
    camera.snap()
animation = camera.animate(interval=(10**-8))
animation.save('fourier.gif')
```
